```
# 02 - Anlise e Visualizao
import pandas as pd
import matplotlib.pyplot as plt

# Carregar dataset salvo como Parquet
input_path = "/dbfs/FileStore/products/products_cleaned.parquet"
df = pd.read_parquet(input_path)

# Estatsticas descritivas
display(df.describe())

# Produtos por categoria - Top 10
top_n = 10
top_categories = df['Category'].value_counts().nlargest(top_n)

plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='cornflowerblue')
plt.title(f'Top {top_n} Categorias com Mais Produtos')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.xticks(rotation=45, ha='right')  # Inclina melhor os rtulos
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Preo mdio por categoria
avg_price = df.groupby('Category')['ListPrice'].mean().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
avg_price.plot(kind='barh', color='seagreen')
plt.title('Preo Mdio por Categoria')
plt.xlabel('Preo Mdio')
plt.ylabel('Categoria')
plt.grid(True)
plt.tight_layout()
plt.show()
```
